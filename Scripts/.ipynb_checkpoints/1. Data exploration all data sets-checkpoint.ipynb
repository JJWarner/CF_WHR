## Consistency Check on DF_WHO

```python
#Column Check
DF_WHO.head(5)
```

```python
#Checking data types
DF_WHO.dtypes
```

```python
# Checking for mixed type data
for col in DF_WHO.columns.tolist():
  weird = (DF_WHO[[col]].map(type) != DF_WHO[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (DF_WHO[weird]) > 0:
    print (col)
```

No issues found

```python
#descriptive analysis WHO
DF_WHO.describe()
```

```python
# Missing Value Checks
DF_WHO.isnull().sum()
```

```python
# Missing Value subset for WHO
DF_WHOnan=DF_WHO[DF_WHO['Age_and_Sex_Population'].isnull()==True]

#Check
DF_WHOnan.head(5)
```
