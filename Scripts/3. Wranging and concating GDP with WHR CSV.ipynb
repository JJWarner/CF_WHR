### creating county mapping

```python
country_mappings = {
    'Afghanistan, Rep. of.': 'Afghanistan',
    'Congo, Democratic Republic of': 'Congo (Kinshasa)',
    'Congo, Republic of': 'Congo (Brazzaville)',
    "Côte d'Ivoire": 'Ivory Coast',
    'North Cyprus':'Cyprus',
    'Northern Cyprus':'Cyprus',
    'Hong Kong S.A.R., China':'Hong Kong',
    'Hong Kong SAR':'Hong Kong',
    'Iran, Islamic Republic of': 'Iran',
    'Korea': 'South Korea',
    'Kyrgyz Republic': 'Kyrgyzstan',
    "Lao People's Democratic Republic": 'Laos',
    'Macedonia, Former Yugoslav Republic of': 'Macedonia',
    'Slovak Republic': 'Slovakia',
    'South Sudan':'Sudan',
    'Syrian Arab Republic': 'Syria',
    'Taiwan':'Taiwan Province of China',
    'Trinidad and Tobago':'Trinidad & Tobago',
    'Yemen, Republic of': 'Yemen',
    'Puerto Rico':'United States'
}
```

```python
# setting unmatched countries in GDP to map
DF_GDP['Country'] = DF_GDP['Country'].replace(country_mappings)
```

```python
# setting unmatched countries in WHR to map
DF_WHR['Country'] = DF_WHR['Country'].replace(country_mappings)
```

```python
#renaming GDP column year lower case
DF_GDP = DF_GDP.rename(columns={'Year': 'year'})
```

# Joining WHR and GDP

```python
# joining all datasets 
DF_all= pd.merge(DF_WHR, DF_GDP, on=['Country', 'year'], how='inner', indicator=True)

display(DF_all)
```

```python
# checking left merge counts
DF_all['_merge'].value_counts()
```

```python
# joining all datasets on left merge for quality check 
DF_allleft= pd.merge(DF_WHR, DF_GDP, on=['Country', 'year'], how='left', indicator=True)

display(DF_allleft)
```

```python
#Isolating left merge counts to determine which countries did not match with GDP
left_merges = DF_all[DF_all['_merge'] == 'left_only']

left_merges.head(47)
```

```python
left_merges['Country'].value_counts()
```
