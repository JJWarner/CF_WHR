# 5. Cleaning DF_all 

```python
# joining all datasets 
DF_all= pd.merge(DF_WHR, DF_GDP, on=['Country', 'year'], how='left', indicator=True)

display(DF_all)
```

# 4. Cleaning WHR_GDP

```python
#renaming GDP (current US$) to GDp for standardization
DF_all=DF_all.rename(columns={"GDP (current US$)":"GDP"})
```

### Adressing Missing values

```python
# Check for missing values
DF_all.isnull().sum()
```

```python
#Identifying missing values
DF_all1 = DF_all[DF_all.isna().any(axis=1)]
print(DF_all1)
```

```python
#creating sub-set of United Arab Emirates rows

corruption = DF_all.loc[DF_all['Country']=='United Arab Emirates']
print(corruption)
```

```python
#Impute missing values with median from United Arab Emirates sub-set 

DF_all['perceptions_of_corruption'].fillna(corruption['perceptions_of_corruption'].median(), inplace=True)

print(DF_all['perceptions_of_corruption'].isnull().sum())
```

```python
#Imputing 0 for missing GDP values

DF_all['GDP'] = DF_all['GDP'].replace(np.nan, 0)

# print the DataFrame
print(DF_all['GDP'].isnull().sum())
```

```python
#Imputing 0 for missing GDP_per_capita values                25

DF_all['GDP_per_capita'] = DF_all['GDP_per_capita'].replace(np.nan, 0)

# print the DataFrame
print(DF_all['GDP_per_capita'].isnull().sum())
```

```python
#Imputing 0 for missing GDP_per_capita_annual_ growth values

DF_all['GDP_per_capita_annual_ growth'] = DF_all['GDP_per_capita_annual_ growth'].replace(np.nan, 0)

# print the DataFrame
print(DF_all['GDP_per_capita_annual_ growth'].isnull().sum())
```

```python
# Check for missing values
DF_all.isnull().sum()
```

```python
#Dropping recently created merge column
DF_all=DF_all.drop(columns=['_merge'])
```
