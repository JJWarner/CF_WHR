# Addressing Outliers

```python
# creating a box plot for 'Reorted_GDP' to review outliers
fig = sns.boxplot(df, y='Reported_GDP')
```

```python
# Running basic descriptives on 'Reported_GDP' column
df['Reported_GDP'].describe()
```

From the descriptive statistics above for the 'Reported_GDP' column  the  mean is 553, but the STD is 2114 with a max value of 21404. 

```python
#creating a function to find outliers using IQR

def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
#printing outliers for Reported_GDP outliers using IQR

outliers = find_outliers_IQR(df['Reported_GDP'])

print('number of outliers:' + str(len(outliers)))
print('max outlier value:'+ str(outliers.max()))
print('min outlier value:' + str(outliers.min()))
```

Ninety-six outliers were identified in this dataset. Isolating all outliers would be equal to approximately 12.7% of the entirety of the data. Given the impact this would have on the overall analysis, outliers cannot be removed, nor can they be capped or replaced due to the value they represent.

Reviewing the box plot above, a small number of outliers appear to have the greatest impact on the overall skew of the data. Outliers will be isolated to the top 10% of the 96 outliers (which is about 28 values seen below), which equates to only 3.7% of the total dataset affected when they are isolated for analysis.

```python
#creating a function to find outliers using IQR /isolating top 5% of IRQ outliers

def find_outliers_IQR(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.90)
   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
#printing outliers for Reported_GDP outliers using IQR

outliers = find_outliers_IQR(df['Reported_GDP'])

print('number of outliers:' + str(len(outliers)))
print('max outlier value:'+ str(outliers.max()))
print('min outlier value:' + str(outliers.min()))
```
