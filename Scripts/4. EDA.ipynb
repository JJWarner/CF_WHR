# 2. Correlations and Heatmaps

```python
df.columns
```

```python
#Creating a subset of only values required for EDA analysis

sub = df[['happiness_score','perceptions_of_economy', 'social_support', 'healthy_life_expectancy',
       'perceptions_of_freedom', 'perceptions_of_corruption','perceptions_of_generosity', 'year', 'GDP', 'GDP_per_capita',
       'GDP_per_capita_annual_ growth'
         ]]
```

```python
# Correlation Grid with Pandas outlining all correlations
sub.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub.corr(), annot = True, cmap=sns.diverging_palette(20, 220, n=200), ax = ax) 

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

plt.savefig("seabornheatmapcorr.png") 
```

```python
#Creating a subset of only values required for EDA analysis

sub2 = df[['happiness_score','perceptions_of_economy','GDP', 'GDP_per_capita',
       'GDP_per_capita_annual_ growth'
         ]]
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(sub2.corr(), annot = True, cmap=sns.diverging_palette(20, 220, n=200), ax = ax) 

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);

plt.savefig("seabornheatmapcorr2.png", bbox_inches = 'tight') 
```

reviewing the heatmap a few variables stand out. Firstly happiness rank has nearly all red squares indicating a negative correlation, which is expected as a lower rank indicated a more positive outcome to the listed variable.
Looking at indicators of economy -'perceptions_of_economy', 'GDP', 'GDP_per_capita','GDP_per_capita_annual_ growth'
'perceptions_of_economy' from the World Happiness report has a strong positive correlation to the happiness score as well as 'GDP_per_capita'. The first represents the participant's perception of economy while the later represent numerical data gathered by the World Bank. 
The other indicators of economy do not show strong correlations to other variables, but spatial analysis may provide more understanding of these variables geographically at later stes.


# 3. Scatter Plots and Pairplots

```python
# Create a pair plot for sub_2

g = sns.pairplot(data=sub,
    y_vars=['perceptions_of_economy','GDP', 'GDP_per_capita', 'GDP_per_capita_annual_ growth'],
    x_vars=['happiness_score','perceptions_of_economy', 'social_support', 'healthy_life_expectancy',
       'perceptions_of_freedom', 'perceptions_of_corruption','perceptions_of_generosity','GDP', 'GDP_per_capita',
       'GDP_per_capita_annual_ growth']
        )
g.savefig("Correlationalchart.png", bbox_inches = 'tight')
```

The purpose of this analysis is to draw insights between reported happiness indicators and GDP.
The indicators 'perceptions_of_economy' and 'GDP_per_capita' exhibit positive correlations with both the happiness score and healthy life expectancy. 'GDP_per_capita' involves raw data comparisons among countries with significantly diverse values and as highlighted in earlier steps, the standard deviation is approximately 2-3 times that of the mean which may accont for the variance contributing to a distinct upward trend in most scatterplots in this row.

These variables stand out due to their diverse correlational connections with other happiness indicators so they will be utilized for ongoing exploration.

```python
# Scatterplot for "GDP_per_capita" & "happiness_score"

g=sns.lmplot(x = "GDP_per_capita", y = "happiness_score", data = df)
g.savefig("scatterplot_gdppercapita_happinessscore")
plt.title('GDP per Capita & Happiness')
```

# 4. Categorical Plots

```python
# cat plot gdp per capita and happiness_score and region

sns.set(style="ticks")
g = sns.catplot(x="GDP_per_capita", 
                y="happiness_score", 
                hue='region',
                palette=diverging_colors,
                data=df)
plt.title('')
g.savefig("Catplt_gpd_and_happiness.png")
```

```python
# bar plot gdp per capita and happiness score

sns.set(style="ticks")
g = sns.barplot(x='region', 
                y="GDP_per_capita", 
                palette=diverging_colors,
                data=df)

g.set_xticklabels(
    g.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.xlabel('')

plt.title('Average GDP per Capita by Region')
plt.savefig("barplt_gdp_region.png", bbox_inches = 'tight')
```

```python
# bar plot gdp per capita and happiness score

sns.set(style="ticks")
g = sns.barplot(x='region', 
                y="happiness_score", 
                palette=diverging_colors,
                data=df)

g.set_xticklabels(
    g.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
plt.xlabel('')

plt.title('Average Happiness Score by Region')
plt.savefig("barplt_happiness_region.png", bbox_inches = 'tight')
```
