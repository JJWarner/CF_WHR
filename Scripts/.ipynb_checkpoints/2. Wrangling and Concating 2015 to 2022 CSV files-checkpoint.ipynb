# Standardizing Column Names and Dropping Columns

## 2015

### Renaming Columns

```python
# Renaming 2015 columns
DF_2015.rename(
    columns={"County":"country", "Region":"region", "Happiness Rank":"happiness_rank", "Happiness Score":"happiness_score", "Family":"social_support", "Economy (GDP per Capita)":"logged_GDP_per_capita", "Health (Life Expectancy)":"healthy_life_expectancy", "Freedom":"freedom_to_make_life_choices", "Generosity":"generosity", "Trust (Government Corruption)":"perceptions_of_corruption"},
    inplace=True,
)

DF_2015
```

### Dropping Columns 

```python
 DF_2015=DF_2015.drop(columns=['Dystopia Residual','standard_error'])

DF_2015.head(3)
```

### Checking Data Types

```python
#Checking data types
DF_2015.dtypes
```

```python
# Checking for mixed type data
for col in DF_2015.columns.tolist():
  weird = (DF_2015[[col]].map(type) != DF_2015[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (DF_2015[weird]) > 0:
    print (col)
```
